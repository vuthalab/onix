```python
%matplotlib ipympl

import numpy as np
import matplotlib.pyplot as plt
from scipy.special import kn, iv
```

```python
sigma = 5e-4
m = 1

omegas = np.logspace(-10, 1, 1000000)
v_bar_sq = omegas / m - 2
omega_bar = omegas / (m * sigma**2)

#A = (5 * np.pi / 32) * (v_bar_sq / sigma**2)**4 * np.exp(-v_bar_sq / sigma**2) * np.heaviside(v_bar_sq, 0)
log_A = np.log(5 * np.pi / 32) + 4 * (np.log(v_bar_sq) - 2 * np.log(sigma)) - v_bar_sq / sigma**2
log_A[v_bar_sq < 0] = -np.inf
B = 2 * np.abs(omega_bar) * kn(1, np.abs(omega_bar))

S_over_tau = sigma**4 * np.exp(log_A) + B
```

```python
fig, ax = plt.subplots()
ax.plot(omegas, S_over_tau)
ax.set_xscale("log")
ax.set_yscale("log")
ax.set_ylim(1e-100, 10)
ax.set_xlabel("$\\omega$ (natural unit)")
ax.set_ylabel("$S(\\omega) / \\tau$")
plt.show()
```
