The position of the SMA fiber input to the Ocean Optics FX spectrometer is kept approximately the same.

```python
import numpy as np
import matplotlib.pyplot as plt
import os

%matplotlib ipympl
```

```python
folder = "C:\\Users\\onix\\oceanview\\data_2024_06_17_spectrometer"
data = {}
for kk in os.listdir(folder):
    full_path = os.path.join(folder, kk)
    data[kk.split("_")[0]] = np.transpose(np.loadtxt(full_path, delimiter="\t", skiprows=14))
```

```python
fig, ax = plt.subplots()
for kk in data:
    if kk != "bkg":
        ax.plot(data[kk][0], data[kk][1] - data["bkg"][1], label=kk)
ax.legend()
plt.show()
```
