```python
def phase_fit(phi, A, B, C, phi_0):
    return A * np.cos(phi - phi_0) +  B * phi + C

def get_phase_fitter(phases, heights, height_errs):
    fitter = Fitter(phase_fit)
    fitter.set_data(phases, heights, height_errs)
    fitter.set_bounds("A", 0, np.inf)

    # setting this bound makes weird fits for certain phi_0's because they want to fit to a value outside the bound
    # fitter.set_bounds("phi_0", 0, 2*np.pi)
    fitter.fit()
    return fitter
```

```python
def get_normalized_transmission(data_number):
    data, header = get_experiment_data(data_number)
    if "fid" in header["params"]["detect"] and header["params"]["detect"]["fid"]["use"]:
        sample_rate = header["params"]["digitizer"]["sample_rate"]
        fid_params = header["params"]["detect"]["fid"]
        start_time = (fid_params["pump_time"] + fid_params["wait_time"]).to("s").magnitude + 7e-6
        fft_end_time = start_time + fft_end_time_s
        times = np.arange(len(data["transmissions_avg"][0])) / sample_rate
        mask = (times > start_time) & (times < fft_end_time)
        time_resolution = times[1] - times[0]
        fs = np.fft.rfftfreq(len(data["transmissions_avg"][0][mask]), d=time_resolution)
        duration = times[mask][-1] - times[mask][0]
        N = duration / time_resolution
        freqs_to_probe = []
        center_detuning = (header["params"]["detect"]["fid"]["probe_detuning"] - header["params"]["detect"]["detunings"][0]).to("Hz").magnitude
        if header["params"]["field_plate"]["use"]:
            electric_field_shift = header["params"]["field_plate"]["stark_shift"].to("Hz").magnitude
            freqs_to_probe.append(center_detuning - electric_field_shift)
            freqs_to_probe.append(center_detuning + electric_field_shift)
        else:
            freqs_to_probe.append(center_detuning)
        freqs_to_probe = np.abs(freqs_to_probe)
        detunings_MHz = np.array(freqs_to_probe) * 1e-6
        indices_to_probe = []
        for freq in freqs_to_probe:
            closest_index = np.argmin(np.abs(fs - freq))
            indices_to_probe.append(closest_index)
        normalized_avg = [[] for kk in freqs_to_probe]
        for kk, d in enumerate(data["transmissions_avg"]):
            if kk >= fft_first_N:
                continue
            ys = np.fft.rfft(d[mask]) * 2 / N
            for ll, closest_index in enumerate(indices_to_probe):
                normalized_avg[ll].append(np.sum(np.abs(ys[closest_index - fft_average_scan: closest_index+fft_average_scan+1])))
        normalized_avg_avg = np.abs(np.average(normalized_avg, axis=1))
        # normalized_avg_std = np.std(np.abs(np.array(normalized_avg) - normalized_avg_avg), axis=1)
        # normalized_avg_std /= len(normalized_avg[0])
        normalized_avg = {"3": -unumpy.uarray(normalized_avg_avg, np.zeros(len(normalized_avg_avg))), "6": ufloat(0,0)}

    else:
        detunings_MHz = header["detunings"].to("MHz").magnitude
        transmissions_avg, transmissions_err = group_and_average_data(data["transmissions_avg"], header["params"]["detect"]["cycles"], return_err=True)
        monitors_avg, monitors_err = group_and_average_data(data["monitors_avg"], header["params"]["detect"]["cycles"], return_err=True)
        normalized_avg = {}
        for kk in transmissions_avg:
            if transmissions_avg[kk].ndim >= 1:
                normalized_avg[kk] = unumpy.uarray(
                    transmissions_avg[kk] / monitors_avg[kk],
                    np.sqrt(
                        (transmissions_err[kk] / monitors_avg[kk]) ** 2
                        + (transmissions_avg[kk] * monitors_err[kk] / monitors_avg[kk]) ** 2
                    )
                )
            else:
                normalized_avg[kk] = ufloat(
                    transmissions_avg[kk] / monitors_avg[kk],
                    np.sqrt(
                        (transmissions_err[kk] / monitors_avg[kk]) ** 2
                        + (transmissions_avg[kk] * monitors_err[kk] / monitors_avg[kk]) ** 2
                    )
                )
    return detunings_MHz, normalized_avg, header
```

```python
def get_results(data_identification):
    data_list = data_identification_to_list(data_identification)
    headers = []
    results = []
    for data_number in data_list:
        header, pop_other_state, total_pop = get_experiment_result(data_number)
        headers.append(header)
        results.append(pop_other_state / total_pop)
    return headers, np.array(results)
```

```python
def averaging_ys(xs, ys):
    xs_unique = np.unique(xs)
    ys_to_avg = [[] for _ in range(len(xs_unique))]
    for x, y in zip(xs, ys):
        ys_to_avg[np.argwhere(xs_unique == x)[0][0]].append(y)
    ys_avg = []
    ys_std = []
    for y_list in ys_to_avg:
        ys_avg.append(np.average(y_list))
        ys_std.append(np.std(y_list))
    ys_avg = np.array(ys_avg)
    ys_std = np.array(ys_std)
    
    return xs_unique, ys_avg, ys_std
```

```python
def get_ramsey_fit(data_range, plot = False, phi_cutoff = np.pi / 2):
    """
    Fits the Ramsey fringe. Returns the RF HSH parameters and the SNR.
    Returns 0 SNR is the fitted phi (once you move it to be within [-pi, pi]) is outside of phi_cutoff.
    """
    headers, results_temp = get_results(data_range)
    mask = np.ones(len(headers), dtype=bool)
    for kk, header in enumerate(headers):
        sequence = header["params"]["sequence"]["sequence"]
        if header["params"]["field_plate"]["amplitude"] < 0:
            mask[kk] = True
        for name, repeat in sequence:
            if name.startswith("lf") and not name.endswith("piov2"):
                index.append(int(name.split("_")[-1]))
                break
                
    amp =  headers[0]["params"]["rf"]["amplitude"]
    t_0 =  headers[0]["params"]["rf"]["T_0"].to("ms").magnitude
    t_e = headers[0]["params"]["rf"]["T_e"].to("ms").magnitude
    t_ch = headers[0]["params"]["rf"]["T_ch"].to("ms").magnitude
    
    xs = np.array([header["params"]["lf"]["phase_diffs"][index[kk]] for kk, header in enumerate(headers)])
    if plot:
        fig, ax = plt.subplots()
    fitted_amp = []
    fitted_amp_unc = []
    
    for kk in range(0, 2):
        ys = unumpy.nominal_values(results_temp[:, kk])
        errs = unumpy.std_devs(results_temp[:, kk])
        xs_avg, ys_avg, errs_avg = averaging_ys(xs[mask], ys[mask])
        
        if plot:
            if kk == 0:
                lambda_label = "$\\lambda = +1$"
            else:
                lambda_label = "$\\lambda = -1$"
            ax.scatter(xs_avg, ys_avg, label=label + lambda_label, marker='o', linestyle='')
            #print(len(xs_avg), len(ys_avg), len(errs))
            #ax.errorbar(xs_avg, ys_avg, errs, fmt = "o", label = label + lambda_label)
        try:
            fitter = get_phase_fitter(xs_avg, ys_avg, None)
            fitted_amp.append(fitter.results['A'])
            fitted_amp_unc.append(fitter.errors['A'])
            xs_fit = np.linspace(np.min(xs_avg), np.max(xs_avg), 100)
            ys_fit = fitter.fitted_value(xs_fit)
            if plot: ax.plot(xs_fit, ys_fit)
        except:
            pass
            
        if plot:    
            if kk == 0:
                try:
                    label_str = label + " $\\lambda = +1$ \n" + "\n".join(fitter.all_results_str().split('\n')[:-2])
                    #label_str = label + " $\\lambda = +1$ \n" + fitter.all_results_str()
                except:
                    label_str = label + " $\\lambda = +1$ \n" + "No Fit." 
                ax.text(0,1.02, label_str, transform = ax.transAxes)
            else:
                try:
                    label_str = label + " $\\lambda = -1$ \n" + "\n".join(fitter.all_results_str().split('\n')[:-2])
                    #label_str = label + " $\\lambda = -1$ \n" + fitter.all_results_str()
                except:
                    label_str = label + " $\\lambda = -1$ \n" + "No Fit." 
                ax.text(0.3,1.02, label_str, transform = ax.transAxes)
    try:
        total_amp = np.mean(fitted_amp)
        total_unc = np.sqrt(sum([i**2 for i in fitted_amp_unc]))
        SNR = total_amp / total_unc
        if plot: ax.text(0.6,1.02, f"SNR = {round(SNR,2)}", transform = ax.transAxes)
    except:
        if plot: ax.text(0.6,1.02, f"SNR = No Fit.", transform = ax.transAxes)
        else: pass
        
    #print(fitted_amp, fitted_amp_unc)
    if plot:
        ax.set_xlabel("Ramsey phase difference (rad)")
        ax.set_ylabel("absorption - offset (arb. unit)")
        ax.text(0.8,1.02,f"#{data_range[0]} - #{data_range[1]}" , transform = ax.transAxes)

        rf_params = f"Amplitude = {amp} \n" + "$T_{ch}$" + f" = {round(t_ch,3)} ms \n $T_0$ = {t_0} ms \n $T_e$ = {t_e} ms"
        ax.text(0.4,0.5,rf_params , transform = ax.transAxes)
        ax.legend()
        ax.grid()
        plt.tight_layout()
        plt.show()
    try:
        phi0 = fitter.results["phi_0"]
        phi0 = (np.pi + phi0) % (2 * np.pi) - np.pi
        if np.abs(phi0) < phi_cutoff:
            return amp, t_0, t_e, t_ch, SNR
        else:
            return amp, t_0, t_e, t_ch, 0
    except:
        return amp, t_0, t_e, t_ch, 0
```

```python
def two_dimensional_grid(x_quantity: str,
                         y_quantity: str,
                         x_value_masks: list,
                         y_value_masks: list,
                         target:str,
                         results: list = results,
                         col_indices: dict = col_indices
                        ):
    """
    Plot a two dimensional grid when doing parameter optimization. Set two parameters. Plot 
    - results: numpy array where results[i] is the result of the ith experiment
    - x_quantity: name of the column for the x quantity, for example "rf_amplitude"
    - x_value_masks: list of masks for this quantity
    - target: entry of results you want to plot for each set of parameters, for example "SNR" or "contrast".
    """
    y_axis = []
    x_axis = []
    vals = []
    
    for y_value_mask in y_value_masks:
        vals_y_fixed = []
        for x_value_mask in x_value_masks:
            try:
                x_value = results[x_value_mask&y_value_mask, col_indices[x_quantity]][0]
                y_value = results[x_value_mask&y_value_mask, col_indices[y_quantity]][0]
                if x_value not in x_axis:
                  x_axis.append(x_value)
                vals_y_fixed.append(np.average(results[x_value_mask&y_value_mask, col_indices[target]]))
            except:
                print("skipping")
                continue
        vals.append(vals_y_fixed)
        y_axis.append(y_value)
    fig, ax = plt.subplots()
    im = ax.imshow(vals, aspect='auto')
    
    # Show all ticks and label them with the respective list entries
    ax.set_xticks(np.arange(len(x_axis)), labels=[round(i, 3) for i in x_axis])
    ax.set_yticks(np.arange(len(y_axis)), labels=[round(i, 3) for i in y_axis])
    
    # Rotate the tick labels and set their alignment.
    plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
             rotation_mode="anchor")
    
    # Loop over data dimensions and create text annotations.
    for i in range(len(x_axis)):
        for j in range(len(y_axis)):
            text = ax.text(i, j, round(vals[j][i], 3),
                           ha="center", va="center", color="w")
    ax.set_xlabel(x_quantity)
    ax.set_ylabel(y_quantity)
    ax.set_title(target)
    ax.text(0,1.02, f"#{results[:,col_indices['first_data_number']][0]} - #{results[:,col_indices['last_data_number']][-1]}", transform = ax.transAxes)
    fig.tight_layout()
    plt.show()
```

```python
def two_dimensional_grid_other_params_fixed(x_quantity: str,
                                            y_quantity: str,
                                            x_value_masks: list,
                                            y_value_masks: list,
                                            param_1_mask: list,
                                            param_2_mask: list,
                                            target: str,
                                            results: list = results,
                                            col_indices: dict = col_indices):
    """
    For use in plotting a grid of subplots in which each subplot is a 2 dimensional grid. 
    Essentially the same code as two_dimensional_grid, although it now applies the the param_1_mask and param_2_mask to hold params 1 and 2 constant 
    as we vary the x and y quantities. 

    Returns the 2 dimensional data as well as x and y axes. 
    """
    y_axis = []
    x_axis = []
    vals = []
    
    for y_value_mask in y_value_masks:
        vals_y_fixed = []
        for x_value_mask in x_value_masks:
            try:
                x_value = results[x_value_mask&y_value_mask&param_1_mask&param_2_mask, col_indices[x_quantity]][0]
                y_value = results[x_value_mask&y_value_mask&param_1_mask&param_2_mask, col_indices[y_quantity]][0]
                if x_value not in x_axis:
                  x_axis.append(x_value)
                vals_y_fixed.append(np.average(results[x_value_mask&y_value_mask&param_1_mask&param_2_mask, col_indices[target]]))
            except:
                print("skipping")
                continue
        vals.append(vals_y_fixed)
        y_axis.append(y_value)
    return vals, x_axis, y_axis
```

```python
def four_dimensional_grid(
     x_quantity: str,
     y_quantity: str,
     x_value_masks: list,
     y_value_masks: list,
     param_1: str,
     param_2: str,
     target:str,
     results: list = results,
     col_indices: dict = col_indices
    ):
    """
    Generate a grid of subplots, where each subplot is a 2 D grid. Each subplot will be the x_quantity versus y_quantity grid. 
    Subplots will differ by the other two parameters, which will be held constant. 
    Each subplot will use the same scaling for the colorbar. The minimum value for the colorbar is set to 3 sigma below target mean. The max is 3 sigma above.
    """
    param_1_values = np.unique(results[:, col_indices[param_1]])
    param_2_values = np.unique(results[:, col_indices[param_2]])
    param_1_masks = [results[:, col_indices[param_1]] == kk for kk in param_1_values]
    param_2_masks = [results[:, col_indices[param_2]] == kk for kk in param_2_values]
    target_mean = np.mean(results[:,col_indices[target]])
    target_std = np.std(results[:,col_indices[target]])
    target_min_value = target_mean - 3 * target_std
    target_max_value = target_mean + 3 * target_std
    
    fig, ax = plt.subplots(len(param_1_values), len(param_2_values), sharex = True, sharey = True)
    fig.suptitle(target)
    #t_0 = param1
    #t_e = param2
    for p, q in enumerate(param_1_values):
        for r, s in enumerate(param_2_values):
            vals, x_axis, y_axis = two_dimensional_grid_other_params_fixed(x_quantity, y_quantity, x_value_masks, y_value_masks, param_1_masks[p], param_2_masks[r], target, results, col_indices)
            im = ax[p,r].imshow(vals, vmin=target_min_value, vmax=target_max_value, cmap='jet', aspect='auto')
            #im = ax[p,r].imshow(vals, cmap='jet', aspect='auto')
            ax[p,r].set_xticks(np.arange(len(x_axis)), labels=[round(i, 3) for i in x_axis])
            ax[p,r].set_yticks(np.arange(len(y_axis)), labels=[round(i, 3) for i in y_axis])
            
            # Rotate the tick labels and set their alignment.
            plt.setp(ax[p,r].get_xticklabels(), rotation=45, ha="right",
                     rotation_mode="anchor")
            
            ax[3,0].set_xlabel(x_quantity)
            ax[3,0].set_ylabel(y_quantity)
            
            ax[p,r].text(0,1.02, f"{param_1} = {q}, {param_2} =  {s}", transform = ax[p,r].transAxes)
            #fig.tight_layout()
            #plt.show()
    plt.colorbar(im)
    plt.show()
```

## LF Rabi Flopping and RF Rabi Flopping HSH Optimization

```python
def rabi_oscillation(t, A, W, B):
    return A * (np.cos(2*pi*W * t))**2 + B
    
def get_rabi_amplitude(data_range, plot = False):
    headers, results_temp = get_results(data_range)
    times = np.array([i.to("us").magnitude for i in headers[0]["params"]["lf"]["durations"]])
    t_axis = np.linspace(np.min(times), np.max(times), int(1e3))
    ys = unumpy.nominal_values(results_temp)[:,0]
    errs = unumpy.std_devs(results_temp)[:,0]

    amp =  headers[0]["params"]["rf"]["HSH"]["amplitude"]
    t_0 =  headers[0]["params"]["rf"]["HSH"]["T_0"]
    t_e = headers[0]["params"]["rf"]["HSH"]["T_e"]
    t_ch = headers[0]["params"]["rf"]["HSH"]["T_ch"]

    try:
        fitter = Fitter(rabi_oscillation)
        fitter.set_data(times, ys, errs)
        fitter.set_p0([0.2,2520e-6,-0.25])
        fitter.fit()
    except:
        pass
    if plot:
        fig, ax = plt.subplots()
        ax.errorbar(times, ys, errs, ls = "", marker = "o")
        try:
            ax.plot(t_axis,fitter.fitted_value(t_axis))
        except:
            pass
        ax.set_xlabel("LF Pulse Time (us)")
        ax.set_ylabel(f"Proportional to Population in $| b \\rangle$ + offset")
        ax.text(0, 1.02, f"#{data_range[0]} - #{data_range[1]}", transform = ax.transAxes)
        
        label = f"Amplitude = {amp} \nT_0 = {t_0} \nT_e = {t_e} \nT_ch = {t_ch}"
        try:
            rabi_frequency = f"Rabi Frequency \n= $2 \\pi \\times$ {present_float(fitter.results['W']*1e6, fitter.errors['W']*1e6)} Hz"  
            chi_2 = f"$\\chi^2 = $ {round(fitter.reduced_chi, 2)}"
            SNR_W = f"SNR in $\\Omega_R$= {round(fitter.results['W'] / fitter.errors['W'], 2)}"
        except: 
            rabi_frequency = "No fit."
            chi_2 = "No fit."
            SNR_W = "No fit."
        ax.text(0.5,0.2,rabi_frequency, transform = ax.transAxes)
        ax.text(0.5,0.05,SNR_W, transform = ax.transAxes)
        ax.text(0.2,0.8,label, transform = ax.transAxes)
        ax.grid()
        plt.tight_layout()
        plt.show()

    try:
        if fitter.results['W'] / fitter.errors['W'] < 2000:
            return amp, t_0, t_e, t_ch, np.nan, np.nan, np.nan, np.nan
        else:
            return amp, t_0, t_e, t_ch, fitter.results['A'], fitter.errors['A'], fitter.results['W'], fitter.errors['W']
    except:
        return amp, t_0, t_e, t_ch, np.nan, np.nan, np.nan, np.nan
```

## Long Term Phase Scan

```python
import onix.models.hyperfine as hyperfine
magnetic_field = 223e-4
ground = hyperfine.states["7F0"]
ground._Hamiltonian = ground.H_total(magnetic_field)
e_g, s_g = ground.energies_and_eigenstates()

a_I_dot_n = (s_g[0].dag() * ground._I_x * s_g[0]).tr()
abar_I_dot_n = (s_g[1].dag() * ground._I_x * s_g[1]).tr()
b_I_dot_n = (s_g[2].dag() * ground._I_x * s_g[2]).tr()
bbar_I_dot_n = (s_g[3].dag() * ground._I_x * s_g[3]).tr()
I_a = (abar_I_dot_n - a_I_dot_n) / 2
I_b = (bbar_I_dot_n - b_I_dot_n) / 2

def combine_polarization_data(results):
    def check_equal_and_append(check_from, check_name, append_to):
        check_col = check_from[:, col_indices[check_name]]
        count = np.unique(check_col)
        if len(count) > 1:
            raise Exception(f"{check_name} is not the same.")
        append_to.append(check_col[0])

    results_array = pandas.DataFrame.from_dict(results).to_numpy()
    keys = list(results.keys())
    indices = list(range(len(keys)))
    col_indices = dict(zip(keys, indices))
    data = {
        "f+": [],
        "f-": [],
        "start_time": [],
        "end_time": [],
        "temp": [],
        "phi0+": [],
        "phi0-": [],
    }
    for kk in range(len(results_array) // 2):
        sub_data = results_array[kk * 2: kk * 2 + 2]
        data_indices = []
        for name in results.keys():
            if name not in ["start_time", "end_time", "D", "freq_center", "temp", "phi0"]:
                if name not in data:
                    data[name] = []
                check_equal_and_append(sub_data, name, data[name])
        for ll, datapoint in enumerate(sub_data):
            if ll == 0:
                start_time = datapoint[col_indices["start_time"]]
            if ll == len(sub_data) - 1:
                end_time = datapoint[col_indices["end_time"]]
            if datapoint[col_indices["D"]]:
                f_p = datapoint[col_indices["freq_center"]]
                phi0_p = datapoint[col_indices["phi0"]]
            else:
                f_m = datapoint[col_indices["freq_center"]]
                phi0_m = datapoint[col_indices["phi0"]]


        data["f+"].append(f_p)
        data["f-"].append(f_m)
        data["phi0+"].append(phi0_p)
        data["phi0-"].append(phi0_m)
        data["start_time"].append(start_time)
        data["end_time"].append(end_time)
        data["temp"].append(np.nanmean(sub_data[:, col_indices["temp"]]))

    data["f+"] = np.array(data["f+"])
    data["f-"] = np.array(data["f-"])
    data["phi0+"] = np.array(data["phi0+"])
    data["phi0-"] = np.array(data["phi0-"])
    data["Z"] = (data["f+"] + data["f-"]) / 4
    data["W_T"] = []
    for kk in range(len(data["f+"])):
        if data["state"][kk] == "a":
            I = I_a
        else:
            I = I_b
        data["W_T"].append((data["f+"][kk] - data["f-"][kk]) / 4 / I)

    keys = list(data.keys())
    indices = list(range(len(keys)))
    col_indices = dict(zip(keys, indices))
    data = pandas.DataFrame.from_dict(data).to_numpy()
    return data, col_indices
```

```python
def cubic_fit(f, f0, a, b):
    return a*(f-f0) + b*(f-f0)**3

def get_cubic_fitter(x, y):
    fitter = Fitter(cubic_fit)
    fitter.set_data(x, y)
    fitter.fit()
    return fitter

def fodof(f, T, D, V,):
    return f*(T-D) + 2*np.arctan((f*np.tan(np.sqrt(4*V**2 + f**2)*D/2))/np.sqrt(4*V**2 + f**2))

def get_fodof_fitter(f, y):
    fitter = Fitter(fodof)
    fitter.set_data(f, y)
    fitter.fit()
    return fitter

def linear_fit(f, f_0, a):
    return a*(f - f_0)

def get_linear_fitter(x, y):
    fitter = Fitter(linear_fit)
    fitter.set_data(x, y)
    fitter.set_p0([2*np.pi, 141146])
    fitter.fit()
    return fitter
```
