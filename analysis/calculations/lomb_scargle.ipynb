```python
%matplotlib ipympl

import numpy as np
import scipy.constants as _c
from scipy.signal import lombscargle
from astropy.timeseries import LombScargle
import matplotlib.pyplot as plt
```

```python
length = 1000
rng = np.random.default_rng()
t_spaces = rng.normal(size=length) * 0.01 + 1
ts = np.cumsum(t_spaces)
ys = np.sin(ts * 2 * np.pi * 0.1)

fig, ax = plt.subplots()
ax.plot(ts, ys)
plt.tight_layout()
plt.show()
```

```python
fs = np.arange(1e-4, 10, 1e-4)
omegas = fs * 2 * np.pi
y_fs = lombscargle(ts, ys, freqs=omegas, precenter=True, normalize=True)

fig, ax = plt.subplots()
ax.plot(fs, y_fs)
ax.set_xscale("log")
plt.tight_layout()
plt.show()
```
