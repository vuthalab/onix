```python
%matplotlib ipympl
import numpy as np
import matplotlib.pyplot as plt

from datetime import datetime
from onix.data_tools import get_persistent_data
```

```python
d, h = get_persistent_data(12, "Magnetometer")
times = [datetime.fromtimestamp(t) for t in d["t"]]
```

```python
def V_to_B(V):
    return V / 0.1 * 0.01

def B_to_V(B):
    return B * 100 * 0.1

fig, ax = plt.subplots()
ax.set_xlabel("time")
ax.set_ylabel("voltage (V)")
ax.plot(times, d["Ch1"], label="x (down)")
ax.plot(times, d["Ch2"], label="y (west)")
ax.plot(times, d["Ch3"], label="z (north)")
ax1 = ax.secondary_yaxis("right", functions=(V_to_B, B_to_V))
ax1.set_ylabel("Magnetic field (G)")
ax.legend()
plt.show()
```
