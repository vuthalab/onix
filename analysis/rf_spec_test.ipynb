```python
import numpy as np
import matplotlib.pyplot as plt

from onix.data_tools import get_experiment_data
```

```python
data, headers = get_experiment_data(6589547)
```

```python
avg_tran_detect_1 = np.average(data["transmission"][()]["detect_1"], axis=0)
avg_tran_detect_2 = np.average(data["transmission"][()]["detect_2"], axis=0)
detunings_MHz = data["detunings_MHz"]

fig, ax = plt.subplots()
ax.scatter(detunings_MHz, avg_tran_detect_1 / avg_tran_detect_2)
plt.show()
```
