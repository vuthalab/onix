```python
import numpy as np
import matplotlib.pyplot as plt

from onix.data_tools import get_experiment_data
```

```python
min_values = []
for kk in range(6589518, 6589568):
    data, headers = get_experiment_data(kk)
    avg_tran_detect_1 = np.average(data["transmission"][()]["detect_1"], axis=0)
    avg_tran_detect_2 = np.average(data["transmission"][()]["detect_2"], axis=0)
    detunings_MHz = data["detunings_MHz"]
    min_values.append(min(avg_tran_detect_1 / avg_tran_detect_2))
    
fig, ax = plt.subplots()
ax.scatter(list(range(len(min_values))), min_values)
plt.show()
```
