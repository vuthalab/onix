```python
%matplotlib ipympl

import numpy as np
import matplotlib.pyplot as plt
from onix.data_tools import get_experiment_data

%config InlineBackend.figure_format = 'retina'
```

```python
#first number: 80792
data_numbers = [80821, 80822]
data = dict([(kk, get_experiment_data(kk)) for kk in data_numbers])
labels = {}
#saved to data[number][1] should be the headers
```

```python
# make the intensity noise plot
fig, ax = plt.subplots(figsize=(10, 8), dpi=100)

for number in data_numbers:
    if "repeats" in data[number][1]['params'].keys():
        ax.plot(data[number][0]["f"], data[number][0]["noise"], label=f"p={data[number][1]['params']['p']}, i={data[number][1]['params']['i']}, d={data[number][1]['params']['d']}, integral limit = {data[number][1]['params']['integral_limit']}, repeats={data[number][1]['params']['repeats']}", alpha=0.8)
    else:
        ax.plot(data[number][0]["f"], data[number][0]["noise"], label=f"p={data[number][1]['params']['p']}, i={data[number][1]['params']['i']}, d={data[number][1]['params']['d']}, integral limit = {data[number][1]['params']['integral_limit']}, repeats = 1", alpha=0.8)

ax.legend()
ax.set_xscale("log")
ax.set_yscale("log")
ax.set_xlabel("Frequency (Hz)")
ax.set_ylabel("Voltage noise density (V/$\\sqrt{\\mathrm{Hz}}$)")
plt.tight_layout()
plt.show()
```

```python
# make the fractional intensity noise plot
fig, ax = plt.subplots()

for number in data_numbers:
    if "repeats" in data[number][1]['params'].keys():
        ax.plot(data[number][0]["f"], data[number][0]["fractional_noise"], label=f"p={data[number][1]['params']['p']}, i={data[number][1]['params']['i']}, d={data[number][1]['params']['d']}, integral limit = {data[number][1]['params']['integral_limit']}, repeats={data[number][1]['params']['repeats']}", alpha=0.8)
    else:
        ax.plot(data[number][0]["f"], data[number][0]["fractional_noise"], label=f"p={data[number][1]['params']['p']}, i={data[number][1]['params']['i']}, d={data[number][1]['params']['d']}, integral limit = {data[number][1]['params']['integral_limit']}, repeats=1", alpha=0.8)
ax.legend()
ax.set_xscale("log")
ax.set_yscale("log")
ax.set_xlabel("Frequency (Hz)")
ax.set_ylabel("Fractional voltage noise density $\\sqrt{\\mathrm{Hz}}^{-1}$")
plt.tight_layout()
plt.show()
```
