```python
%matplotlib ipympl

import numpy as np
import matplotlib.pyplot as plt
from onix.data_tools import get_experiment_data
```

```python
# we want to save the data as "f", "noise", and "fractional_noise"
# inside the params we need the p i and d params
```

```python
data_numbers = []
data = dict([(kk, get_experiment_data(kk)) for kk in data_numbers])
labels = {}
#saved to data[number][1] should be the headers
```

```python
# make the intensity noise plot
fig, ax = plt.subplots()

for number in data_numbers:
    ax.plot(data[number][0]["f"], data[number][0]["noise"], label=f"p={data[number][1]['p']}, i={data[number][1]['i']}, d={data[number][1]['d']}", alpha=0.8)

ax.legend()
ax.set_xscale("log")
ax.set_yscale("log")
ax.set_xlabel("Frequency (Hz)")
ax.set_ylabel("Voltage noise density (V/$\\sqrt{\\mathrm{Hz}}$)")
plt.tight_layout()
plt.show()
```

```python
# make the fractional intensity noise plot
fig, ax = plt.subplots()

for number in data_numbers:
    ax.plot(data[number][0]["f"], data[number][0]["fractional_noise"], label=f"p={data[number][1]['p']}, i={data[number][1]['i']}, d={data[number][1]['d']}", alpha=0.8)

ax.legend()
ax.set_xscale("log")
ax.set_yscale("log")
ax.set_xlabel("Frequency (Hz)")
ax.set_ylabel("Fractional voltage noise density $\\sqrt{\\mathrm{Hz}}^{-1}$")
plt.tight_layout()
plt.show()
```
