```python
%matplotlib inline

import datetime
import numpy as np
import matplotlib.pyplot as plt
from onix.analysis.debug.laser_linewidth import LaserLinewidth
from onix.data_tools import get_persistent_data
```

```python
data, headers = get_persistent_data(3, "Pulse Tube")
times = [datetime.datetime.fromtimestamp(kk) for kk in data["times"]]
```

```python
fig, ax = plt.subplots()
ax.plot(times, data["T_coolant_in_C"].T[1], label="T_in")
ax.plot(times, data["T_coolant_out_C"].T[1], label="T_out")
ax.plot(times, data["T_oil_C"].T[1], label="T_oil")
ax1 = ax.twinx()
ax1.set_ylabel("state")
ax1.plot(times, data["compresser_states"].astype(int), label="state", color="C7", zorder=0)
ax1.set_yticks([0, 1])
ax1.set_yticklabels(["off", "on"])
ax.set_xlabel("Time")
ax.set_ylabel("Temperature (C)")
ax.legend()
plt.tight_layout()
plt.show()
```
