```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 5000
M = 200
ss = np.zeros((M, N))

def dist_uniform():
    a = np.random.randint(1, 31)
    b = np.random.randint(1, 33)
    return np.random.uniform(-a, -a+b, N)
    
def dist_triangular():
    a = np.random.randint(1, 55)
    b = np.random.randint(1, 33)
    return np.random.triangular(-a, (-a+b)+15, -a+b+33, N)
    
for m in range(M):
    for f in [dist_uniform, dist_triangular]:
        s = dist_triangular()
        ss[m] = s
        count, bins, ignored = plt.hist(s, 20, density=True, alpha=0.2)
plt.show()
```

```python
t = np.sum(ss, axis=0)
count, bins, ignored = plt.hist(t, 50, density=True)
plt.show()
```
