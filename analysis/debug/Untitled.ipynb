```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 1000
M = 100
ss = np.zeros((M, N))

for m in range(M):
    s = np.random.uniform(-1, 1, N)
    ss[m] = s
    count, bins, ignored = plt.hist(s, 15, density=True, alpha=0.2)
plt.show()
```

```python
t = np.sum(ss, axis=0)
count, bins, ignored = plt.hist(t, 50, density=True)
plt.show()
```
