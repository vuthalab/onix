```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("/home/onix/Desktop/119MHz_matched.csv", delimiter=",", skiprows=1).T
```

```python
fig, ax = plt.subplots()
ax.plot(data[0] / 1e6, data[1])
ax1 = ax.twinx()
unwrapped_angles = data[2].copy()
for kk in range(len(unwrapped_angles)):
    if unwrapped_angles[kk] < unwrapped_angles[kk + 1] - 100:
        unwrapped_angles[kk + 1:] -= 360
        break
ax1.plot(data[0] / 1e6, unwrapped_angles, color="C1")
ax1.set_ylabel("Phase (deg)")
ax.set_xlabel("Frequency (MHz)")
ax.set_ylabel("Reflection coeffcient (dB)")
plt.show()
```
